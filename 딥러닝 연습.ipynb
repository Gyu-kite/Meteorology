# 최소제곱법 예제

```python
x= np.array([2,4,6,8])
y= np.array([81,93,91,97])
```

```python
mx= float(np.mean(x)) # x의 평균
my= float(np.mean(y)) # y의 평균
```

```python
type(mx)
```

```python
divisor =  sum([(i-mx)**2 for i in x]) # 최소제곱법의 분모
print(divisor)
```

```python
def top (x, mx, y, my):
  d=0
  for i in range(len(x)):
    d = d + (x[i]-mx)*(y[i]-my) # 최소제곱법의 분자
  return d
dividend = top(x, mx, y, my)
```

```python
a = dividend / divisor # 기울기 구하기
a
```

```python
b = my -(mx*a) # y절편 구하기
b
```

```python
print("기울기 =",a)
print("y 절편 =",b)
```

# 평균 제곱 오차
최소제곱법으로 기울기와 y절편을 구했지만 이 공식만으로는 모든 상황을 해결하기는 어렵다.

### *그러나 여러개의 입력을 처리하기에는 무리가 있다.*
앞선 상황은 변수가 한개지만 나중에는 여러개 나옴

기울기와 절편을 다르게 구하는 방법은 조금씩 수정해 나가는 방법!

오차가 최소가 될 때까지 이 과정을 계속 반복하는 방법
따라서 선의 오차를 평가하는 방법이 필요하다.  오차를 구할때 가장 많이 사용되는  방법이다.

### *오차 = 실제값 - 예측값*

```python
fake_a = 3
fake_b = 76

def predict(x):
  return fake_a*x+fake_b

predict_result = []
```

```python
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  x1=x[i]
  y1=y[i]
  pre=predict(x[i])
  print("공부시간=",x1,"실제점수=",y1,"예측점수=",pre)
```

```python
predict_result
```

```python
# 평균제곱오차고익
n=len(x)
def mse(y, y_pred):
  return (1/n)*sum((y-y_pred)**2)
```

```python
print("평균제곱오차 :" +str(mse(y,predict_result)))
```

# 선형회귀모델 먼저 긋고 수정하기
## 경사하강법

경사하강법은 미분 기울기르 사용하여 오차가 가장 작은 기울기를 찾아내는것
오차와 기울기의 함수에서 도함수의 기울기가 0인 지점

절편을 구할 때도  같은 방법으로 구할 수 있음


```python
y_pred = a*x+b # 예측값을 구하는 식.
error = y-y_pred # 식제 값과 비교한 오차를 error로 놓는다.

a_diff = (2/n)*sum(-x*(error)) # 오차함수를 a 로 편미분한 값
b_diff = (2/n)*sum(-(error)) # 오차함수를 b 로 편미분한 값
```

```python
lr =0.03 # 학습률을 정합니다.
a = a - lr * a_diff # 학습률을 곱해 기존의 a 값을 업데이트 한다.
b = b - lr * b_diff # 학습률을 통해 기존의 b 값을 업데이트 한다.
```

## 선형회귀모델 실습 

```python
import numpy as np
import matplotlib.pyplot as plt

# 공부시간 x 와 성적 y 의 넘파이 배열을 만든다.
x= np.array([2,4,6,8])
y= np.array([81,93,91,97])

# 데이터의 분포를 그래프로 나타낸다.
plt.scatter(x,y)
plt.show()
```

```python
a=0
b=0
```

```python
lr =0.03 # 학습률을 정합니다.
epochs = 2001
n=len(x)
for i in range(epochs):
  y_pred = a*x+b # 예측값을 구하는 식.
  error = y-y_pred # 식제 값과 비교한 오차를 error로 놓는다.

  a_diff = (2/n)*sum(-x*(error)) # 오차함수를 a 로 편미분한 값
  b_diff = (2/n)*sum(-(error)) # 오차함수를 b 로 편미분한 값


  a = a - lr * a_diff # 학습률을 곱해 기존의 a 값을 업데이트 한다.
  b = b - lr * b_diff # 학습률을 통해 기존의 b 값을 업데이트 한다.

  if i % 100==0:
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f" %(i,a,b))

y_pred = a*x+b # 예측값을 구하는 식.

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show
```

## 텐서플로우 옵티마이저



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 공부시간 x 와 성적 y 의 넘파이 배열을 만든다.
x= np.array([2,4,6,8])
y= np.array([81,93,91,97])

model=Sequential()

model.add(Dense(1,input_dim=1,activation='linear'))
model.compile(optimizer='sgd',loss='mse') # 오차 수정을 위해 경사하강법sgd, 오차의 정도를 판단하기 위해 mse사용 평균제곱오차
model.fit(x,y,epochs=2000)
```

```python
# 데이터의 분포를 그래프로 나타낸다.
plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우의 예상점수는 %.02f점입니다." %(hour, prediction))
```

```python
#다중선형회귀 

x= np.array([[2,0],[4,4],[6,2],[8,3]])
y= np.array([81,93,91,97])

model=Sequential()

model.add(Dense(1,input_dim=2,activation='linear')) # 입력해야하는 변수가 2개이므로 인풋=2
model.compile(optimizer='sgd',loss='mse') # 오차 수정을 위해 경사하강법sgd, 오차의 정도를 판단하기 위해 mse사용 평균제곱오차
model.fit(x,y,epochs=2000)
```

```python
hour = 7
private_class  = 4
prediction = model.predict([[hour,private_class]]) #차원이 2이므로 [[]]
print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다." %(hour,private_class, prediction))
```

## 로지스틱 회귀

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,8,10,12,14])
y = np.array([0,0,0,1,1,1,1])

model=Sequential()

model.add(Dense(1,input_dim=1,activation='sigmoid'))
model.compile(optimizer='sgd',loss='binary_crossentropy') # 오차 수정을 위해 경사하강법sgd, 오차의 정도를 판단하기 위해 mse사용 평균제곱오차
model.fit(x,y,epochs=5000)

```

```python
# 데이터의 분포를 그래프로 나타낸다.
plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%점입니다." %(hour, prediction*100))
```
